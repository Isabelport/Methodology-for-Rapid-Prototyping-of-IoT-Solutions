```python
import plotly.express as px
import os
import pandas as pd
import numpy as np
import plotly.graph_objects as go

#import csv
pd.set_option('display.max_rows', 300)
```

```python
# Assuming you have a start date and an end date in the format 'DD/MM/YYYY'
dataset = "rfid"

all_data = {}

def read_files(start_date, end_date):
    # Convert the start and end dates to datetime objects
    start_date = pd.to_datetime(start_date, format='%d/%m/%Y')
    end_date = pd.to_datetime(end_date, format='%d/%m/%Y')

    # Create a date range
    date_range = pd.date_range(start=start_date, end=end_date)

    # Iterate over the date range
    for date in date_range:
        day = str(date.day).zfill(2)
        month = str(date.month).zfill(2)
        date = day + "_" + month
        all_data = process_file_and_stack(date)


    
    return all_data

def process_file_and_stack(date):

    
    folder_path = dataset + '_' + date + '/'
    for station in range(1, 4):
        file_path = os.path.join(folder_path, f'rfid_{date}_{station}.csv')
        
        # Check if the file exists
        if os.path.exists(file_path):
            # Read the CSV file
            data_station = pd.read_csv(file_path, parse_dates=['ts'])

            # Remove first empty column if it exists
            if data_station.shape[1] == 3:
                data_station = data_station.drop(data_station.columns[0], axis=1)
            # Remove entries with 'gabarit' equal to 0
            data_station= data_station[data_station['gabarit'] != 0]

            # Concatenate or initialize the DataFrame in the dictionary
            if station in all_data:
                all_data[station] = pd.concat([all_data[station], data_station])
            else:
                all_data[station] = data_station

            all_data[station] = all_data[station].reset_index(drop=True)

    return all_data

def process_test_file():
    folder_path = 'rfid_fabrica'
    file_path = os.path.join(folder_path, 'rfid_final.csv')

    df = pd.read_csv(file_path, parse_dates=['Stamp'], dayfirst=True)
    if df.shape[1] == 4:
        df = df.drop(df.columns[3], axis=1)
        df = df.drop(df.columns[1], axis=1)
    df.to_csv(file_path, index = False)

    return df


start_date = '25/09/2023'
end_date = '09/10/2023'
#all_data = read_files(start_date, end_date)
all_data[4] = process_test_file()
#display(all_data[4])
```
